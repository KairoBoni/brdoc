```python
text +="{\n"
text += '"docs" : ['
for line in lines:
    text+="{\n"
    line = line.split("|")
    text +='"CRM":"'+line[0]+'",\n'
    text +='"nome":"'+line[1]+'",\n'
    text +='"qtd": 0,\n'
    text += '"avlist": [{avaliacao: ""}],\n'
    if(line[2] == 'I'):
        text +='"situacao":"INATIVO",\n'
    else:
        text +='"situacao":"Ativo",\n'
    text +='"nota" : 0,\n'
    if(line[2] == "I"):
        text +='"dataSaida":"'+line[0]+'",\n'
    else:
        text +='"dataSaida":null,\n'
    
    if(line[5].replace(" ", "") == ""):
        text +='"cidade":null,\n'
        text +='"estado":null,\n'
    else:
        text +='"cidade":"'+line[5]+'",\n'
        text +='"estado":"'+line[6]+'",\n'
    if(line[7] == '\n'):
        text +='"esp":"NÃO POSSUI"\n'
    else:
        line[7] = line[7].replace("\n", "")
        text +='"esp":"'+line[7]+'"'
    
    if(line[0] == "053537"):
        text+="}\n"
    else:
        text+="},\n"
        

text += ']'
text +="}"

```

```python
import json

t = json.loads(text)

json.dump(t, jsonT)

jsonT.close()
```
